```python
import numpy as np
import pandas as pd
import os
import glob
import warnings
warnings.simplefilter('ignore')
data_list=glob.glob('/Users/akash/Downloads/Painting Dataset/Raw Data/painting/*.jpg')
np.array(data_list).shape
```

```python
from tensorflow.keras.preprocessing.image import load_img
load_img(data_list[20])
```

```python
#Grayscaling it cz it differentiates between the shadow details and the highlights of an image because it is mainly in 2 spatial dimensions (2D) rather than 3D.
#resizing it to 64x64
#The resulting 'paint_dataset' DataFrame contains the label column and the image data.
import pandas as pd
image_list=[]
for i in range(1361):
    image_list.append(load_img(data_list[i], target_size=(64,64),grayscale=True))
img_data=[]
for k in range(1360):
    img_data.append(np.array(image_list[k]).flatten())
zeros=pd.DataFrame(np.zeros(1360).astype(int))
img_data_pd=pd.DataFrame(img_data)
paint_dataset=pd.concat([zeros,img_data_pd],axis=1)
paint_dataset
```

```python
#Grayscaling it cz it differentiates between the shadow details and the highlights of an image because it is mainly in 2 spatial dimensions (2D) rather than 3D.
#resizing it to 64x64
#The resulting 'paint_dataset' DataFrame contains the label column and the image data.
image_list2=[]
for i in range(3747):
    image_list2.append(load_img(data_list2[i], target_size=(64,64),grayscale=True))
img_data2=[]
for k in range(3747):
    img_data2.append(np.array(image_list2[k]).flatten())
ones=pd.DataFrame(np.ones(3747).astype(int))
img_data_pd2=pd.DataFrame(img_data2)
photo_dataset=pd.concat([ones,img_data_pd2],axis=1)
photo_dataset
```

```python
from sklearn.model_selection import train_test_split
x =dataset.iloc[:,1:]
y =dataset.iloc[:,[0]]
x_train_pd, x_test_pd, y_train_pd, y_test_pd = train_test_split(x, y, test_size=0.2,random_state=123)
```

```python
#normalizes data from 1 to 0.
x_train = x_train_pd.astype('float32')/255
y_train = y_train_pd.astype('float32')
x_test = x_test_pd.astype('float32')/255
y_test = y_test_pd.astype('float32')
```

```python
#change data into numpy array
x_train= x_train.to_numpy()
y_train= y_train.to_numpy()
x_test= x_test.to_numpy()
y_test= y_test.to_numpy()

#reshape train data and test data into 64 * 64 * 1channel
x_train = x_train.reshape(x_train.shape[0], 64, 64, 1).astype('float32')
x_test = x_test.reshape(x_test.shape[0], 64, 64, 1).astype('float32')
```

```python
#convert y_train and y_test into 2 categories
from tensorflow import keras as kr
y_train = kr.utils.to_categorical(y_train, 2)
y_test = kr.utils.to_categorical(y_test, 2)
num_classes = y_train.shape[1]
num_classes
```

```python
#import libraries
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv2D
from tensorflow.keras.layers import MaxPooling2D
from tensorflow.keras.layers import Dense, Activation, Flatten
from tensorflow.keras.optimizers import Adam

#Build an ordinary "Deep Learning" model with CNN and maxpooling by using Keras.
model = Sequential()
model.add(Conv2D(512, (5, 5), input_shape=(64, 64, 1), activation='relu'))
model.add(MaxPooling2D(pool_size=(2, 2)))
model.add(Flatten())
model.add(Dense(128, activation='relu'))
model.add(Dense(num_classes, activation='softmax'))
#Choose an optimizer and compile the model.
model.compile(optimizer = Adam(learning_rate = 0.01), loss = 'categorical_crossentropy', metrics = ['accuracy'])
#And print the summary of the model.
print(model.summary())
```

```python
#model fitting
model1 = model.fit(x_train, y_train,batch_size=128, epochs=25)
```

```python
import matplotlib.pyplot as plt


history = model1.history

# Metrics to plot
metrics = ['accuracy', 'loss']

plt.figure(figsize=(15, 5))

# Loop through the metrics
for i, metric in enumerate(metrics):
    plt.subplot(1, 2, i+1)
    plt.title(metric)
    # Access the metric from the history dictionary
    plt_values = history[metric]
    plt.plot(plt_values, label='train1')
    plt.legend()

plt.show()
```

```python
import pandas as pd
from sklearn.metrics import roc_auc_score, f1_score, recall_score, precision_score, accuracy_score
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.linear_model import LogisticRegression

# Assuming x and y are your features and labels
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)

# DataFrame to store scores
scores_df = pd.DataFrame(columns=["Model", "Accuracy", "AUC", "F1", "Recall", "Precision"])

# Evaluate each model
for name, model in models.items():
    model.fit(x_train, y_train)
    y_pred = model.predict(x_test)
    y_pred_proba = model.predict_proba(x_test)[:, 1]  # For AUC

    # Calculate metrics
    accuracy = accuracy_score(y_test, y_pred)
    auc = roc_auc_score(y_test, y_pred_proba)
    f1 = f1_score(y_test, y_pred)
    recall = recall_score(y_test, y_pred)
    precision = precision_score(y_test, y_pred)

    # Append to DataFrame
    scores_df = scores_df.append({"Model": name, "Accuracy": accuracy, "AUC": auc, "F1": f1, "Recall": recall, "Precision": precision}, ignore_index=True)

# Print the DataFrame
print(scores_df)
```
